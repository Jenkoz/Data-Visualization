**1.2.2 Fix date_Added**

```python
df[df['date_added'].isna()]
```

```python
 df = df[df['date_added'].notna()]
```

```python
# Replacing null values with 'missing' in country column
df['country'].fillna('Missing',inplace=True)
df.isnull().sum().sum() # check for missing values
```

```python
 '''
The "date_added" feature has the date seperated by spaces.
Which makes it harder to extract the year and month.
'''
df['date_added'] # see the values in the date_added column
```

```python
 # converting into proper date time format
df["date_added"] = pd.to_datetime(df['date_added'])
df["date_added"]
```

```python
'''
As a dateTime object, we can now index just the year or month from the date and
add them as new columns to our dataset
'''
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
'''
An example of an entry where we have multiple countries.
We are going to apply the 'split' function to split the string by ','
'''
two_countries=df['country'].unique()[54]
# By splitting them into a list we can select the first country only
print(two_countries.split(','),"-->",two_countries.split(',')[0])
```

```python
 '''
Filter through the dataset can be done by adding condition in the index
df[CONDITION] --> only datapoints that satisfy the condition will pass through
Example:
df[ df['type']=='TV Show' ] --> selects only datapoints with 'type' set␣
,→as 'TV Show'
'''
fig = px.histogram(x=df[df['type']=='TV Show']['duration'])
#plotting in descending order
fig.update_xaxes(categoryorder="total descending")
#adding title and axis labels
fig.update_layout(title="Distribution of shows duration", # title for the chart
xaxis_title="Duration of the Show", # modify the label for x-axis
)
fig.show()
```

```python
movie_df= df[df['type']=='Movie'] # selecting only movies data
movie_df.dropna(axis=0,inplace=True)# getting rid of empty rows
temp_duration = movie_df['duration'].apply(lambda x: int(x.split(' ')[0]))
temp_duration # this contains the duration of movies in minutes
```

```python
 '''
Access a group of rows and columns by label(s) or a boolean array.
``.loc[]`` is primarily label based, but may also be used with a
boolean array.
'''
# filtering through durations where the number of minutes is less than 90 minutes
movie_df.loc[temp_duration.loc[temp_duration < 90].index, 'duration_bin'] = 'Less than 1.30 hours'
# filtering through durations where the number of minutes is greater than 90 minutes but less than 150
movie_df.loc[temp_duration.loc[(temp_duration >= 90) & (temp_duration < 150)].index, 'duration_bin'] = 'More than 1.30 - less than 2.30 hours'
# filtering through durations where the number of minutes is greater than 150 minutes
movie_df.loc[temp_duration.loc[temp_duration >= 150].index, 'duration_bin'] = 'More than 2.30 hours'
```

```python
movie_df.head(2)
```

```python
fig = px.histogram(x=movie_df['duration_bin'], color=movie_df['duration_bin'])
fig.update_layout(
title="Distribution of movies duration",
xaxis_title="Duration of the Movies",
)
fig.show()
```

```python
fig_stack_bar = px.histogram(df, y='rating', color='type', title='Which has the highest Rating TV shows or Movies?',
color_discrete_sequence=['#b20710', 'orange'])
fig_stack_bar.update_xaxes(showgrid=False)
fig_stack_bar.update_yaxes(showgrid=False, categoryorder='total ascending', ticksuffix=' ', showline=False)
fig_stack_bar.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig_stack_bar.update_layout(legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5))
```

```python
dff = df.copy()
df_tv_show = dff[dff['type']=='TV Show'][['rating', 'type']].rename(columns={'type':'tv_show'})
df_movie = dff[dff['type']=='Movie'][['rating', 'type']].rename(columns={'type': 'movie'})
df_tv_show = df_tv_show.rating.value_counts().reset_index(). rename(columns={'index':'tv_show'})
df_tv_show['rating_final'] = df_tv_show['rating']
# making rating column value negative (multiplying every value by -1)
df_tv_show['rating'] *= -1
df_movie = df_movie.rating.value_counts().reset_index().rename(columns={'index': 'movie'})
```

```python
 df_movie.head(5)
```

```python
#create a subplot with 2 columns and ashared y-axis
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, horizontal_spacing=0)
# bar plot for tv shows
fig.append_trace(go.Bar(x=df_tv_show.rating, y=df_tv_show.tv_show, orientation='h', showlegend=True,
text=df_tv_show.rating_final, name='TV Show', marker_color='#221f1f'), 1, 1)
# bar plot for movies
fig.append_trace(go.Bar(x=df_movie.rating, y=df_movie.movie, orientation='h', showlegend=True, text=df_movie.rating,
name='Movie', marker_color='#b20710'), 1, 2)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False, categoryorder='total ascending', ticksuffix='', showline=False)
fig.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig.add_annotation(dict(x=0.88, y=0.4, ax=0, ay=0,
xref = "paper", yref = "paper",
text= "<b>97%</b> people prefer Movies over TV Shows on <br>Netflix.<br> Large number of people watch TV-MA rating"
))
```

```python
df_month = pd.DataFrame(df.month_added.value_counts()).reset_index().rename(columns={'index':'month','month_added':'count'})
# converting month number to month name using 'replace'
'''
'replace' takes a dict that takes 2 values {val1,val2}
val1 --> original value
val2 --> replacment value
'''
df_month['month_final'] = df_month['month'].replace({1:'Jan', 2:'Feb', 3:'Mar',4:'Apr', 5:'May', 6:'June', 7:'July', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov',12:'Dec'})
df_month
```

```python
fig_month = px.funnel(df_month, x='count', y='month_final', title='Best month for releasing Content',
height=350, width=600, color_discrete_sequence=['#b20710'])
fig_month.update_xaxes(showgrid=False, ticksuffix=' ', showline=True)
fig_month.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
```

```python
count_per_year=df.groupby(['year_added','type']).agg({'show_id': 'count'}).reset_index().rename(columns = {'show_id' : "count"})
vc1 = count_per_year[count_per_year['type']=='TV Show']
vc1 = vc1.sort_values('year_added')
vc2 = count_per_year[count_per_year['type']=='Movie']
vc2 = vc2.sort_values('year_added')
trace1 = go.Scatter(x=vc1['year_added'], y=vc1['count'], name="TV Shows",marker=dict(color="orange"), )
trace2 = go.Scatter(x=vc2['year_added'], y=vc2['count'], name="Movies",marker=dict(color="#b20710"))
data = [trace1, trace2]
fig_line = go.Figure(data)
fig_line.update_traces(hovertemplate=None)
fig_line.update_xaxes(showgrid=False)
fig_line.update_yaxes(showgrid=False)
large_title_format = 'Tv Show and Movies impact over the Year'
small_title_format = "<span style='font-size:13px; font-family:Tahoma'>Due to Covid updatation of content is slowed."
fig_line.update_layout(title=large_title_format + "<br>" + small_title_format, height=400,
hovermode="x unified",
xaxis_title=' ', yaxis_title=" ",
legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5))
fig_line.add_annotation(dict(x=0.8, y=0.1, ax=0, ay=0,
xref = "paper", yref = "paper",
text= "Highest number of <b>Tv Shows</b><br> were released in <b>2019</b><br> followed by 2017."
))
fig_line.add_annotation(dict(x=0.54, y=0.95, ax=0, ay=0,
xref = "paper", yref = "paper",
text= "Highest number of <b>Movies</b> were relased<br> in <b>2019</b> followed by 2020"
))
fig_line.show()
```
